# General Properties of Probability Distributions



Probability distributions have several general properties, including:

1. **Nonnegative probabilities:** All probabilities in a probability distribution must be greater than or equal to 0 and less than or equal to 1.

2. **Sum of probabilities:** The sum of all possible outcomes in a probability distribution must equal 1.

3. **Location, spread, and shape:** These are the three basic properties of a distribution. Location refers to the typical value, such as the mean, while spread refers to how much smaller values differ from larger ones.

4. **Expected value:** The expected value of X is also known as the mean of the distribution f.

5. **Variance:** Variance measures the spread of a random variable around its expectation.

6. **Standard deviation:** The standard deviation is the square root of the variance.

7. **Random variables:** Variables that follow a probability distribution are called random variables.

8. **Probability mass function:** A probability mass function is used for random variables that can take any designated finite or countable values

# Discrete probability distributions

---

A discrete probability distribution shows the probabilities of specific outcomes for a variable that can only take distinct, separate values, like whole numbers.








## **Binomial Distribution**

---

**Definition**

---

The binomial distribution models the number of successes in a fixed number of trials, where each trial has two possible outcomes (success or failure) and a constant probability of success, ùëù.

---

**Key Features**

---

- The binomial distribution is a discrete distribution with a finite number of possibilities.
- When observing a series of what are known as Bernoulli trials, the binomial distribution emerges.
- Arises from Bernoulli trials, where each trial has exactly two outcomes: success or failure.

---

**Examples**

---

- Flipping a coin 5 times and counting the heads..
- Counting defective items in a batch of 20 with a 5% defect rate.
- Tracking how many of 50 customers make a purchase, with a 30% chance per customer.
- Scoring correct answers in a 10-question multiple-choice quiz, with a 25% chance per question.

```python
# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success
```

```python
# Possible outcomes
x = np.arange(0, n + 1)
x
```

```python
# Probability mass function
pmf = binom.pmf(x, n, p)
pmf
```

```python
# Plotting the distribution
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(x)
plt.show()
```


## **Discrete uniform Distribution**

---

**Definition**

---

The discrete uniform distribution models a situation where all possible outcomes have an equal probability of occurring.

---

**Key Features**

---

- Applies to discrete random variables.
- All outcomes are equally likely within the defined range.
- The probability of each outcome is (1 / ùëõ), where
ùëõ is the total number of possible outcomes.


---

**Examples**

---

- The suit of a randomly drawn playing card.
- Picking a random day of the week, where each day has an equal chance.


```python
# Parameters for the discrete uniform distribution
low = 1  # Lower bound (inclusive)
high = 7  # Upper bound (exclusive, equivalent to a die roll)
```

```python
# Possible outcomes
x = np.arange(low, high)
x
```

```python
# Probability mass function (all probabilities are equal)
pmf = [1 / (high - low) for _ in x]
pmf
```

```python
# Plotting the distribution
plt.bar(x, pmf, color='lightgreen', edgecolor='black')
plt.title(f'Discrete Uniform Distribution (low={low}, high={high - 1})')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(x)
plt.show()
```

## **Poisson Distribution**

---

**Definition**

---

The Poisson distribution models the probability of an event occurring ùëò times within a fixed interval of time or space, given that the events happen independently and at a constant average rate.

---

**Key Features**

---

- Applies to count data.
- Events are independent of one another.
- The average rate of occurrence ùúÜ, where Œª is constant over time or space.
- Suitable for rare or random events in a given interval.

---

**Examples**

---

- The number of text messages received in a day.
- Car accidents occurring in a day.
- Calls received by a call center in an hour.
- Typing errors per page in a book.
- Bus arrivals at a stop within 10 minutes.
- Emails received in a day.
- Stars observed in a fixed area of the night sky.

```python
# Parameters for the Poisson distribution
mu = 4  # Average rate of occurrences (lambda)
```

```python
# Possible outcomes
x = np.arange(0, 15)
x
```

```python
# Probability mass function
pmf = poisson.pmf(x, mu)
pmf
```

```python
# Plotting the distribution
plt.bar(x, pmf, color='lightcoral', edgecolor='black')
plt.title(f'Poisson Distribution (mu={mu})')
plt.xlabel('Number of Occurrences')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(x)
plt.show()
```

# Continuous probability distributions

---

A continuous probability distribution describes the probabilities of a continuous random variable's possible values, which can be any value within a specified range. Unlike discrete distributions, where the outcomes are distinct and countable, continuous distributions involve an infinite, uncountable set of values.

- A continuous variable can take any value between its lowest and highest points.
  - For example, time can be considered a continuous variable, where the possible values can range from 1 second to 1 billion seconds, and any value in between.
- The probability of a continuous variable taking any specific value is essentially zero, as the possible values are infinite.
  - However, the probability that the variable falls within a certain range or interval is greater than zero.
- In continuous distributions, probabilities are represented by areas under the curve, as opposed to specific values.

## Normal distribution

---

**Definition**

---

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric around its mean, with data closer to the mean occurring more frequently than data farther away. It has a characteristic bell-shaped curve.

---

**Key Features**

---

- Symmetrical around the mean.
- The mean, median, and mode are equal.
- Most data points lie within three standard deviations of the mean.
- The mean is 0, and variance is finite for the standard normal distribution.
- Often used to model natural phenomena and measurement data.

---

**Examples**

---

- Heights of people in a population.
- Test scores in a standardized exam.
- Daily temperatures in a specific city over time.
- Measurement errors in scientific experiments.


```python
# Parameters for the Normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation
```

```python
# Range of values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
```

```python
# Probability density function
pdf = norm.pdf(x, mu, sigma)
```

```python
# Plotting the distribution
plt.fill_between(x, pdf, color='blue', alpha=0.6, label=f'N({mu}, {sigma}^2)')
plt.title(f'Normal Distribution (mean={mu}, std={sigma})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.legend()
plt.show()
```

## Continuous uniform distribution

---

**Definition**

---

Describes data for which equal-sized intervals have equal probability. The continuous uniform distribution describes a scenario where all outcomes within a given range are equally likely to occur, with equal probability for any interval of the same length.

---

**Key Features**

---

- All outcomes in the specified range have the same probability.
- The probability density function is flat.
- Defined by two parameters: the minimum and maximum values of the range.
- The mean is the midpoint of the range, and the variance is based on the range's width.

---

**Examples**

---

- The amount of time cars wait at a red light.
- The height of a randomly chosen person between 5 and 6 feet.
- The time taken for a computer to process a task within a given time window.


```python
# Parameters for the Uniform distribution
a = -2  # Lower bound
b = 2   # Upper bound
```

```python
# Range of values
x = np.linspace(a - 1, b + 1, 1000)
```

```python
# Probability density function for Uniform distribution
pdf = uniform.pdf(x, a, b - a)
```

```python
# Plotting the area chart
plt.fill_between(x, pdf, color='green', alpha=0.6, label=f'Uniform({a}, {b})')
plt.title(f'Uniform Distribution (a={a}, b={b})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.legend()
plt.show()
```

## Log-normal distribution

---

**Definition**

---

The log-normal distribution describes right-skewed data and is the probability distribution of a random variable whose logarithm is normally distributed.

---

**Key Features**

---

- Data is right-skewed, meaning most values are clustered on the lower end, with a tail extending to the right.
- The variable itself follows a normal distribution when transformed into its logarithmic form.
- Commonly used to model data that cannot be negative and where values increase exponentially.

---

**Examples**

---

- The average body weight of different mammal species.
- The distribution of income or wealth in a population.
- The size of particles in a dust storm.

```python
# Parameters for the Log-normal distribution
mu = 0  # Mean of the underlying normal distribution
sigma = 1  # Standard deviation of the underlying normal distribution
```

```python
# Range of values (x should be positive for log-normal distribution)
x = np.linspace(0, 10, 1000)
```

```python
# Probability density function for Log-normal distribution
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))
```

```python
# Plotting the chart
plt.fill_between(x, pdf, color='orange', alpha=0.6, label=f'Log-normal(Œº={mu}, œÉ={sigma})')
plt.title(f'Log-normal Distribution (Œº={mu}, œÉ={sigma})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Exponential distribution

---

**Definition**

---


Describes data that has higher probabilities for small values than large values. It‚Äôs the probability distribution of time between independent events.

---

**Key Features**

---

- Used to model the time between events in a Poisson process.
- Has a memoryless property, meaning the probability of an event occurring in the future is independent of past events.
- The mean and standard deviation are equal and are defined by the rate parameter
ùúÜ.

---

**Examples**

---

- Time between earthquakes.
- Time between customer arrivals at a service counter.
- The lifespan of a light bulb.

```python
# Parameter for the Exponential distribution
lambda_ = 1  # Rate parameter (1/mean)
```

```python
# Range of values
x = np.linspace(0, 10, 1000)
```

```python
# Probability density function for Exponential distribution
pdf = expon.pdf(x, scale=1/lambda_)
```

```python
# Plotting the chart
plt.fill_between(x, pdf, color='purple', alpha=0.6, label=f'Exponential(Œª={lambda_})')
plt.title(f'Exponential Distribution (Œª={lambda_})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
